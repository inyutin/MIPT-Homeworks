```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('time.csv', header=None)[0]
n = len(data)
data
```

```python
plt.figure(figsize=(16,6))
plt.title('Ускорение программы в зависимости от числа потоков')
plt.plot(np.arange(1, 9), data[0]/data)
plt.xticks(np.arange(0, len(data)+1))
plt.yticks(np.arange(0, max(data[0]/data)+1))
plt.xlabel("Число потоков")
plt.ylabel("Ускорение")
plt.grid(True)
plt.legend()
plt.show()
```
