```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data = pd.read_csv('temp.txt', sep='   ', header=None)
data.head()
```

Визуализируем температуру на пластине

```python
_, ax = plt.subplots(figsize=(16, 12))  
sns.heatmap(data, ax=ax, cmap="RdBu_r", center=0)
```

Сравним реперные точки и некоторые срезы нашей пластины по $Ox$

```python
x = [0.005, 0.025, 0.045, 0.065, 0.085, 0.115, 0.15, 0.1775, 0.225, 0.345, 0.405, 0.435, 0.495]
U = [80, 70, 60, 50, 40, 30, 20, 15, 10, 10, 15, 20, 30]
```

```python
plt.figure(figsize=(16, 12))
plt.figure(figsize=(16,6))
for i in range(10, 90, 20):
    plt.plot(np.arange(0, 0.51, 0.005), data.T[i], label='Срез при y={}'.format(i))
plt.plot(x, U, label='Реперные точки',linewidth=3.0)
plt.plot(x, U, 'o')
plt.grid(True)
plt.xlabel('Координата x');
plt.ylabel('Температура U');
plt.legend()
plt.title('Сравнение реперных точек и температур на срезах', fontsize=14);
```

Изучим ускорение

```python
fast = pd.read_csv('fast.csv',header=None)
slow = pd.read_csv('slow.csv',header=None)

plt.figure(figsize=(16,6))
plt.title('Ускорение программы в зависимости от числа процессов')
plt.plot(np.arange(1, 11), fast.T[0].mean()/fast.T.mean(), label='fast')
plt.plot(np.arange(1, 11), slow.T[0].mean()/slow.T.mean(), label='slow')
plt.xticks(np.arange(0, len(fast)+1))
plt.yticks(np.arange(0, max(fast.T[0].mean()/fast.T.mean())+1))
plt.xlabel("Число процессов")
plt.ylabel("Ускорение")
plt.grid(True)
plt.legend()
plt.show()
```
