```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('time.csv')
n = len(data)
data.head()
```

```python
plt.figure(figsize=(16,6))
plt.plot(np.arange(1, n+1), data['critical'][0]/data['critical'])
plt.title('Ускорение программы, использующая критическую секцию')
plt.xticks(np.arange(0, n+1))
plt.yticks(np.arange(1, max(data['critical'][0]/data['critical'])+1, 0.5))
plt.grid(True)
plt.xlabel("Количество потоков")
plt.ylabel("Ускорение")
plt.show()
```

```python
plt.figure(figsize=(16,6))
plt.plot(np.arange(1, n+1), data['reduction'][0]/data['reduction'])
plt.title('Ускорение программы, использующая reduction')
plt.xticks(np.arange(0, n+1))
plt.yticks(np.arange(1, max(data['reduction'][0]/data['reduction'])+1, 0.5))
plt.grid(True)
plt.xlabel("Количество потоков")
plt.ylabel("Ускорение")
plt.show()
```
