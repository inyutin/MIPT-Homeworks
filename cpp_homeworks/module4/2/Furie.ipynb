```python
import numpy as np
import matplotlib.pyplot as plt
from __future__ import division
import pylab as py
from matplotlib import mlab
%matplotlib inline
```

```python
def FFT(x, m = 1):
    N = len(x)
    if N <= 1: return x
    even = FFT(x[::2], m)
    not_even = FFT(x[1::2], m)

    exponents = np.exp(-2j * m * np.pi * np.arange(N) / N)
    return np.concatenate([even + exponents[ : N // 2] * not_even,
                     even + exponents[N // 2 :] * not_even])
def IFFT(fft):
    x = FFT(fft, -1)
    return x / len(x)
```

```python
def do(f, D = 0):
    print("Количество обнуленных коэффицентов - ", D)
    x = np.linspace(0.1,10.24,128)
    y = f(x)
    fft = FFT(y)
    fft[:-D] = 0
    damaged_y = IFFT(fft)
    figsize=(8, 6)
    plt.title("График оригинальной функции")
    plt.plot(x,y)  
    plt.show()
    
    plt.title("График функции под IFFT")
    plt.plot(x,damaged_y)
    plt.show()
    
    plt.title("Графики в сравнении")
    plt.plot(x, y, label = "График оригинальной функции")
    plt.plot(x, damaged_y, label = "График функции под IFFT")
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    plt.show()
```

```python
def build_function(D = 0):
    print("y = x")
    do(lambda x : x, D)
    
    print("y = x^2")
    do(lambda x: x*x, D)
    
    print("y = sin(x)")
    do(np.sin, D)    
    
    print("y = np.sin(x^2)")
    do(lambda x: np.sin(x*x), D)
    
    print("y = np.sin(x)")
    do(lambda x: (np.sin(x) / x), D)
```
